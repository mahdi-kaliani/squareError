```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
num_features = X.shape[1]
num_classes = 1
learning_rate = 0.1
num_epochs = 100
```

```python
weights = np.random.randn(num_classes, num_features)
bias = np.random.randn(num_classes)
```

```python
errors = []
for epoch in range(num_epochs):
    total_error = 0
    for i in range(X.shape[0]):
        output = np.dot(weights, X[i]) + bias
        y_pred = np.where(output > 0, 1, 0)
        error = y[i] - y_pred
        weights += learning_rate * error * X[i].reshape((num_features, 1)).T
        bias += learning_rate * error
        bias += learning_rate * error
        total_error += np.abs(error)
    errors.append(total_error)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.arange(num_epochs), np.arange(len(errors)))
Z = np.array(errors)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('Epoch')
ax.set_ylabel('Data point')
ax.set_zlabel('Squared error')
plt.show()
```
